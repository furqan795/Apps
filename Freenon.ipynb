```python
import requests

def check_freenom_availability(domains):
    available = []
    base_url = "https://my.freenom.com/includes/domains/fn-check_availability.php"
    extensions = ['.tk', '.ml', '.ga', '.cf', '.gq']

    for name in domains:
        try:
            params = { 'domain': name }
            headers = {
                'User-Agent': 'Mozilla/5.0',
                'Referer': 'https://my.freenom.com/domains.php'
            }
            response = requests.get(base_url, params=params, headers=headers)
            result = response.text.lower()

            for ext in extensions:
                full_domain = name + ext
                if f"{full_domain} is available" in result:
                    available.append(full_domain)
        except Exception as e:
            print(f"Error checking {name}: {e}")

    return available

# ✅ Test it
domain_names = ['bmicalcfast', 'bmiguidehub', 'ezbmicalc', 'mybmistats', 'bmicheckerx']
available_domains = check_freenom_availability(domain_names)

print("\n✅ Available Domains:")
for domain in available_domains:
    print(domain)
```
